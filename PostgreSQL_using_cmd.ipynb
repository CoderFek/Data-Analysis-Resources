# **CONNECT TO DATABASE**

Using the default database "postgres"

```
psql -U postgres
```



Or



```
psql -h localhost -p 5432 -U postgres
```



Note that 'postgres' is a default database in psql which can be used to write queries and create other database and tables. Also, 'localhost' can be used as your default host when there are no ip addresses mentioned and 5432 is the port number which is available by default in case you do not have other ports to perform queries in your database.

-h, -p and -U are just commands used in cmd. \? will give you a list of commands you will be using to perform different operations in your database.

# **CREATING A TABLE**

A table with multiple columns can be created in the following way.

```
CREATE TABLE person (
	id INT NOT NULL,
	first_name VARCHAR(50) NOT NULL,
	last_name VARCHAR(50) NOT NULL,
	email VARCHAR(150),
	gender VARCHAR(50) NOT NULL,
	date_of_birth DATE NOT NULL,
	country_of_birth VARCHAR(50)
);
```
Here, id is the Primary key.


---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Accessing the table.

```
\d person
```

### 1. **SELECT and FROM**



```
SELECT * FROM person;
```



image.png



```
SELECT first_name, last_name, country_of_birth FROM person LIMIT 10;
```



image.png

### 2. **WHERE**



```
SELECT * FROM person WHERE gender = 'Female' LIMIT 10;
```



image.png



```
SELECT * FROM person WHERE gender != 'Female' LIMIT 10;
```



image.png

### 3. **ORDER BY**



```
SELECT * FROM person ORDER BY country_of_birth LIMIT 10;
```



image.png



```
SELECT first_name, last_name, email FROM person ORDER BY email LIMIT 10;
```



image.png

### 4. **DISTINCT**



```
SELECT DISTINCT country_of_birth
FROM person
ORDER BY country_of_birth LIMIT 10;
```



image.png

### 5. **AND / OR**



```
SELECT * FROM person WHERE gender = 'Female'  AND country_of_birth = 'Poland' LIMIT 10;
```



image.png



```
SELECT * FROM person WHERE gender = 'Female'  AND (country_of_birth = 'Poland' OR country_of_birth = 'Brazil') LIMIT 10;
```



image.png

### 6. **IN**



```
SELECT * FROM person WHERE country_of_birth IN ('China','Brazil','France') LIMIT 10;
```



image.png

### 7. **BETWEEN**



```
SELECT * FROM person WHERE date_of_birth BETWEEN '2015-01-01' AND '2023-01-01';
```
